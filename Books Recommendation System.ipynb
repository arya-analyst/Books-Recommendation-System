```python
popularity_df = ratings_count.merge(ratings_avg,  on = ('Book-Title'))
popularity_df.head()
```

### Collaborative Filtering Based Recommendation System

###### We cannot use each and every books and each and every user's input to make this system.

```python
book_ratings.head()
```

```python
book_ratings.groupby('User-ID').count()
```

```python
## We are taking only those users into accout here who have reviewed atleast 100 books.

a = book_ratings.groupby('User-ID').count()['Book-Rating'] > 100
filtered_users = a[a].index
```

```python
filtered_users_list = book_ratings[book_ratings['User-ID'].isin(filtered_users)]
```

```python
## We are taking only those books into accout here which have been reviewed atleast 30 times.

b = filtered_users_list.groupby('Book-Title').count()['Book-Rating'] > 30
filtered_books = b[b].index
```

```python
filtered_books_list = filtered_users_list[filtered_users_list['Book-Title'].isin(filtered_books)]
```

```python
pt = filtered_books_list.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt.isnull().sum()
```

```python
pt.fillna(0, inplace = True)
```

```python
pt.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores[0]
```

```python
similarity_scores.shape
```

```python
def recommend_books(books):
    index = np.where(pt.index == books )[0][0] #fetch index
    similar_books = sorted(list(enumerate(similarity_scores[index])), key = lambda x: x[1], reverse=True)[1:6]

    for i in similar_books:
        print (pt.index[i[0]])
```

```python
## first of all we wil fetch the index of the input i.e., the book name

    # np.where(pt.index == '1984')[0][0]

## fetch the similarity for that vector
    # similarity_score[]

## enumerate helps us in maintaining the index.
## list converts it in a list
## sorted sorts the list and by introducing the lambda function for **key** parameter

    # sorted(list(enumerate(similarity_scores[0])), key = lambda x: x[1], reverse=True)[1:6]

## i iterates through all the 5 most similar books and prints them 1-BY-1
## i[0] only prints the index of the simmilar books - we don't need to show the **similarity_scores**
## passing i[0] in pt.index[] gives us the name of the books

## The name of the books are the **index** in the pivot table.
```

```python
recommend_books('1984')
```
